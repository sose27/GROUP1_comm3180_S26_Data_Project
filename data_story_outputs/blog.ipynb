## Demographic and Socioeconomic County Profiles

But what do these Republican-voting counties actually look like? Using a collection of data from the American Community Survey, we sought to explore this by constructing demographic profiles for three of them — Placer, Kern, and Kings - examining their racial composition, dominant industries of employment, and median household income. We chose counties that are meaningfully distinct from one another to better understand the multiple dimensions of California conservatism. [^1] 

[^1]: The Hispanic racial cateogry was measured separately from other the racial categories, so the racial percentages do not add up to 100%.

### Placer County

![](data/placer_county_profile.png)

Placer County is the most affluent of the three, with a median household income of $109,375 and a median age of 42 - both well above California averages. The county is predominantly white at around 85%, with Hispanic and Asian residents each making up roughly 11-12%. Its employment base is dominated by health care, retail, and accommodation and food services, reflecting a suburban service economy.

### Kern County

![](data/kern_county_profile.png)

Kern County looks very different. Its median household income is $63,883 and its population is more racially mixed - roughly 70% white with a Hispanic population of around 55%. Sitting on one of the largest oil fields in the country, it makes sense that agriculture and mining is one of the most dominant industries in the county. Construction also features more prominently here than in Placer, which reflects a more working class economic base.

### Kings County

![](data/kings_county_profile.png)

Ostensibly, Kings County is quite similar to Kern — a comparable income ($68,540), a similarly young median age of 32, a large Hispanic population of around 57%, and a significant share of the work force concentrated in agriculture and mining. However, Kings’ economy centers more heavily on food production and processing which completely reorients their political priorities.

What these three profiles reveal is that Republican voting across California's inland counties is not rooted in a single demographic or economic reality. Placer is wealthy, predominantly white, and suburban - a more traditional fit for Republican affiliation. Kern and Kings, on the other hand, complicate that picture. Both are majority-minority counties with lower incomes and economies built around agriculture and physical labor, demographics that in many parts of the country would predict Democratic alignment. The fact that they vote Republican suggests that economic identity and industry shape political behavior in ways that cut across racial and income lines. Ultimately, the distinctions between these counties reveal just how varied the coalition behind California’s Republican vote actually is.

## Proposition Voting and Political Complexity

The demographic and economic profiles above show that Republican-voting counties in California are not politically identical. Some are wealthy and suburban, while others are younger, more working-class, and tied to agriculture or industry.

To better understand how counties respond to different political issues, we also examined county-level proposition voting from the 2024 California election. Unlike presidential elections, propositions isolate opinions on specific policy questions rather than party loyalty or candidate appeal. Because every county votes on the same statewide propositions, ballot measures give us a clearer way to compare political preferences across California.

We focused on three propositions that each represent a different dimension of politics:

- **Proposition 3** addressed same-sex marriage and broader questions surrounding civil rights and social liberalism.
- **Proposition 32** proposed raising the minimum wage, capturing attitudes toward economic redistribution, labor protections, and government intervention in the economy.
- **Proposition 36** increased penalties for certain drug and theft crimes, reflecting views on criminal justice, public safety, and punitive policy approaches.

Together, these measures allow us to compare how counties respond to social, economic, and criminal justice issues separately rather than assuming counties hold one fixed ideological identity.

### Why Proposition Voting Matters

### What Are Propositions
California propositions are statewide ballot measures that allow voters to decide policy questions directly.
Unlike candidate elections, propositions show how counties vote on specific issues rather than political parties alone.

### Why Use Them
Because every county votes on the same issue, propositions let us compare political preferences across California without the noise of candidate personality, incumbency, or campaign dynamics. They show what counties actually support when voting on policy itself.

### Focus Measures
##### Prop 3 (Same-Sex Marriage): Captures attitudes on social liberalism and civil rights, particularly around LGBTQ+ rights and the role of government in protecting individual freedoms.
##### Prop (Minimum Wage Increase): Reflects views on economic policy and government intervention, highlighting support for labor protections and redistribution across different types of counties.
##### Prop 36(Crime and Drug Penalties): Represents preferences on criminal justice and public safety, capturing how counties differ in support for punitive versus reform-oriented approaches to crime.

### Together, these propositions allow us to compare how counties respond to social, economic, and criminal justice issues separately, rather than assuming a single, consistent political identity.
