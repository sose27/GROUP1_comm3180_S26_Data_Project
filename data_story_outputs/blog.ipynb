# How Blue is California?

### A County-Level Exploration of Voting Patterns

## Introduction

As students who are all from California, we have grown up hearing the same narrative repeated over and over: California is a solidly “blue” state. In national media and political discourse, it is often treated as politically uniform, with a strong and consistent Democratic identity. 

But from personal experience, this characterization has always felt somewhat incomplete. California is an incredibly large and diverse state, with major differences between coastal cities, suburban areas, and inland rural regions. This raised an important question for us: *is California actually as politically uniform as it is often portrayed, or is there more variation beneath the surface?*

To begin exploring this question, we first turned to county-level voting data. By looking at how different counties vote, rather than just statewide results, we can get a more detailed picture of political behavior across the state.

## County-Level Voting Patterns

We began by visualizing voting margins across California counties. The map below shows how Democratic and Republican support varies geographically.

![](data/heat_map.png)

*Figure 1: County-level voting margins across California*

At first glance, the map already challenges the idea of a uniformly “blue” state. While coastal regions show strong Democratic margins, inland and more rural counties consistently lean Republican. 

This suggests that California’s political identity is more geographically divided than it might appear from statewide results alone. However, a map can only show broad patterns, it does not fully capture how voting behavior changes over time or how competitive different regions are.

## Economic Factors and Voting Patterns

To better understand what drives these differences, we examined how economic variables relate to voting behavior.

One of the strongest relationships we found was with income. The graph below shows a clear negative relationship between median household income and Republican vote share, meaning that higher-income counties tend to vote less Republican.

![](data/income.png)

We also looked at industry, particularly the percentage of people employed in agriculture and mining. Here, we see a positive relationship: counties with more agriculture-based employment tend to vote more Republican.

This suggests that economic structure—especially the distinction between urban and rural economies—plays a key role in shaping political behavior.

![](data/agriculture.png)

## Demographic Factors and Voting Patterns

In addition to economic variables, we also explored demographic factors.

The graph below shows the relationship between racial composition and Republican vote share. Counties with higher percentages of white residents tend to vote more Republican, indicating a moderate positive relationship.

![](data/race.png)

We also examined median age, but found that it has only a weak relationship with voting patterns. This suggests that age alone does not strongly explain differences across counties.

![](data/age.png)

Overall, these findings highlight that demographic composition, alongside economic factors, helps explain the variation in voting behavior across California.
