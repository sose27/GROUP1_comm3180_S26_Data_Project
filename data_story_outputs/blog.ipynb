# How Blue is California?

### A County-Level Exploration of Voting Patterns

## Introduction

During election cycles, the national media tends to characterize California as a solidly “blue” and politically uniform state.

But as students who all grew up in California, this characterization has always felt somewhat incomplete to us. While it's true California hasn't voted for a Republican president in almost 40 years, the state is incredibly large and diverse, with major differences between coastal cities, suburban areas, and inland rural regions. This raised an important question for us: *is California actually as politically uniform as it is often portrayed, or is there more variation beneath the surface?*

To explore this question, we drew on county-level presidential election results, demographic data, and ballot measure voting records. By comparing how counties vote alongside their demographic and socioeconomic differences, we can construct a more nuanced picture of political behavior across the state.

## Demographic and Socioeconomic County Profiles

But what do these Republican-voting counties actually look like? Using a collection of data from the American Community Survey, we sought to explore this by constructing demographic profiles for three of them — Placer, Kern, and Kings - examining their racial composition, dominant industries of employment, and median household income. We chose counties that are meaningfully distinct from one another to better understand the multiple dimensions of California conservatism. [^1] 

[^1]: The Hispanic racial cateogry was measured separately from other the racial categories, so the racial percentages do not add up to 100%.

### Placer County

![](data/placer_county_profile.png)

Placer County is the most affluent of the three, with a median household income of $109,375 and a median age of 42 - both well above California averages. The county is predominantly white at around 85%, with Hispanic and Asian residents each making up roughly 11-12%. Its employment base is dominated by health care, retail, and accommodation and food services, reflecting a suburban service economy.

### Kern County

![](data/kern_county_profile.png)

Kern County looks very different. Its median household income is $63,883 and its population is more racially mixed - roughly 70% white with a Hispanic population of around 55%. Sitting on one of the largest oil fields in the country, it makes sense that agriculture and mining is one of the most dominant industries in the county. Construction also features more prominently here than in Placer, which reflects a more working class economic base.

### Kings County

![](data/kings_county_profile.png)

Ostensibly, Kings County is quite similar to Kern — a comparable income ($68,540), a similarly young median age of 32, a large Hispanic population of around 57%, and a significant share of the work force concentrated in agriculture and mining. However, Kings’ economy centers more heavily on food production and processing which completely reorients their political priorities.

What these three profiles reveal is that Republican voting across California's inland counties is not rooted in a single demographic or economic reality. Placer is wealthy, predominantly white, and suburban - a more traditional fit for Republican affiliation. Kern and Kings, on the other hand, complicate that picture. Both are majority-minority counties with lower incomes and economies built around agriculture and physical labor, demographics that in many parts of the country would predict Democratic alignment. The fact that they vote Republican suggests that economic identity and industry shape political behavior in ways that cut across racial and income lines. Ultimately, the distinctions between these counties reveal just how varied the coalition behind California’s Republican vote actually is.

## Proposition Voting and Political Complexity

The demographic and economic profiles above show that Republican-voting counties in California are not politically identical. Some are wealthy and suburban, while others are younger, more working-class, and tied to agriculture or industry. But demographics and economics only tell part of the story.

To better understand how counties respond to different political issues, we also looked at county-level proposition voting from the 2024 California election. Unlike presidential elections, propositions isolate opinions on specific policy questions rather than party loyalty or candidate appeal. Because every county votes on the same statewide propositions, ballot measures give us a clearer way to compare political preferences across California.

We focused on three propositions that each represent a different dimension of politics:

- **Proposition 3** addressed same-sex marriage and broader questions surrounding civil rights and social liberalism.
- **Proposition 32** proposed raising the minimum wage, capturing attitudes toward economic redistribution, labor protections, and government intervention in the economy.
- **Proposition 36** increased penalties for certain drug and theft crimes, reflecting views on criminal justice, public safety, and punitive policy approaches.

Together, these measures allow us to compare how counties respond to social, economic, and criminal justice issues separately rather than assuming counties hold one fixed ideological identity.

### Counties Vote Differently Depending on the Issue

The maps above reveal major differences in how counties vote depending on the policy area being considered.

Support for **Proposition 3** is strongest in coastal and urban counties, particularly in the Bay Area, reflecting stronger support for socially liberal policies and civil rights protections. Inland and more rural counties show noticeably lower levels of support.

Support for **Proposition 32**, however, extends beyond major Democratic urban centers. While coastal counties still show strong support, several more working-class and inland counties also supported raising the minimum wage, suggesting broader cross-county agreement on certain economic issues even among counties that often vote Republican in presidential elections.

In contrast, **Proposition 36** performs strongest in inland and rural counties. Counties in the Central Valley and northern interior California showed particularly high support for tougher crime penalties, reflecting a more conservative approach to criminal justice and public safety policy.

Taken together, these results demonstrate that county-level politics in California cannot be reduced to a simple liberal-versus-conservative divide. Counties that may vote Republican in presidential elections do not necessarily oppose economic intervention, while Democratic counties may still diverge on issues related to crime and public safety.

#### These proposition patterns also help contextualize the county profiles examined earlier.

Placer County, for example, reflects a more traditional suburban Republican coalition: wealthier, older, and predominantly white. Its voting patterns align more closely with conservative positions on economic and criminal justice issues while remaining less extreme than many rural inland counties.

Kern and Kings Counties complicate that picture. Despite being majority-Hispanic and lower-income counties — demographic characteristics often associated with Democratic voting nationally — both counties still leaned strongly Republican in the 2024 presidential election. At the same time, proposition voting suggests these counties are not uniformly conservative across every issue area.

This tension reinforces one of our central findings: California’s inland Republican coalition is not built from a single demographic profile. Economic structure, industry, geography, and issue salience all shape political behavior differently across counties.

### What Proposition Voting Adds to the Story

The proposition results reinforce our broader argument that California’s political identity is more complicated than its statewide blue reputation suggests. Presidential election results show whether counties lean Democratic or Republican overall, but proposition voting shows where those leanings become more complicated.

Rather than representing fixed political identities, county voting behavior shifts depending on the specific issue being debated. A county may be conservative in presidential elections but still support certain worker protections, or Democratic overall but less enthusiastic about criminal justice reform. This issue-specific variation helps explain why California contains multiple political landscapes within one state.
