### Research Overview

#### Research Question


Which demographic and socioeconomic factors are most strongly associated with Republican vote share across California counties?

#### Dependent Variable

Republican vote share across California counties (2024 presidential election)


#### Independent Variable

-Demographics

    -Race: % White (non-Hispanic), % Hispanic, % Black
    -Median age
    
-Socioeconomic

    -Education: % Bachelor’s degree or higher
    -Income: median household income
    -Unemployment rate
    -Employment by industry
    
-Political

    -Proposition voting results
